```
import cv2
```

```
img = cv2.imread('/content/hulu3.jpeg')
```

```
import numpy as np
```

```
import PIL
```

```
import matplotlib.pyplot as plt
```

```
plt.imshow(img)
```

```
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```
# Find variance of laplacian and check against a threshold to see if it is blurred
img_blurred_val = cv2.Laplacian(img, cv2.CV_64F).var()
print(img_blurred_val)
```

```
img = cv2.GaussianBlur(img, (5, 5), 1) # Blur image for better canny edge detection
```

```
img_blurred_val = cv2.Laplacian(img, cv2.CV_64F).var()
print(img_blurred_val)
```

```
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.imsave('/content/imgcanny.jpg', img_rgb)
```

```
non_th_canny = cv2.Canny(img, 100, 200)
```

```
img_rgb = cv2.cvtColor(non_th_canny, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.imsave('/content/imgcanny.jpg', img_rgb)
```

```
img_rgb = cv2.cvtColor(non_th_canny, cv2.COLOR_BGR2RGB)
plt.imsave('/content/imgcanny.jpg', img_rgb)
```

```
threshed_img = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 21, 15)
```

```
img_rgb = cv2.cvtColor(threshed_img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```
img = cv2.GaussianBlur(threshed_img, (5, 5), 1) # Blur image for better canny edge detection
img_rgb = cv2.cvtColor(threshed_img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```
th_canny = cv2.Canny(threshed_img, 100, 200)
```

```
img_rgb = cv2.cvtColor(th_canny, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```
img_rgb = cv2.cvtColor(th_canny, cv2.COLOR_BGR2RGB)
plt.imsave('/content/imgcanny.jpg', img_rgb)
```
